Задание 1
Импортируйте библиотеку Pandas и дайте ей псевдоним pd. Создайте датафрейм authors со столбцами author_id и author_name, в которых соответственно содержатся данные: [1, 2, 3] и ['Тургенев', 'Чехов', 'Островский'].
Затем создайте датафрейм book cо столбцами author_id, book_title и price, в которых соответственно содержатся данные:  
[1, 1, 1, 2, 2, 3, 3],
['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
[450, 300, 350, 500, 450, 370, 290].

```python
import pandas as pd
import numpy as np
authors = pd.DataFrame({
    'author_id': [1, 2, 3],
    'author_name': ['Тургенев', 'Чехов', 'Островский']
})
book = pd.DataFrame({
    'author_id': [1, 1, 1, 2, 2, 3, 3],
    'book_title': ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
    'price': [450, 300, 350, 500, 450, 370, 290]
})
book
```

Задание 2
Получите датафрейм authors_price, соединив датафреймы authors и books по полю author_id.

```python
authors_price = pd.merge(authors, book, on = 'author_id', how = 'right')
authors_price
```

Задание 3
Создайте датафрейм top5, в котором содержатся строки из authors_price с пятью самыми дорогими книгами.b

```python
top5 = authors_price.sort_values(by = 'price', ascending = False, inplace = False).head(5)
top5
```

Задание 4
Создайте датафрейм authors_stat на основе информации из authors_price. В датафрейме authors_stat должны быть четыре столбца:
author_name, min_price, max_price и mean_price,
в которых должны содержаться соответственно имя автора, минимальная, максимальная и средняя цена на книги этого автора.


```python
authors_groupby = authors_price.groupby('author_name')['price'].agg({'min', 'max', 'mean'})
#authors_groupby
authors_stat = pd.DataFrame({#'author_name': authors_groupby.index,
                             'min_price':authors_groupby['min'],
                             'max_price':authors_groupby['max'],
                             'mean_price':authors_groupby['mean']
                            }).reset_index()
display(authors_price)
display(authors_stat)
```

Используйте для этого функцию pd.pivot_table. При этом столбцы должны называться "твердая" и "мягкая", а индексами должны быть фамилии авторов. Пропущенные значения стоимостей заполните нулями, при необходимости загрузите библиотеку Numpy.

```python
book_info = authors_price.pivot_table('price', index = 'author_name', columns = 'cover', aggfunc = 'sum').fillna(0.0)
display(book_info)
display(type(book_info.iloc[1,1]))
display(type(book_info.iloc[0,1]))
```

Ознакомьтесь с описанием данных и скачайте датасет netflix_titles.csv с сайта Kaggle.com https://www.kaggle.com/shivamb/netflix-shows
Считайте датасет. Посмотрите, сколько наблюдений и сколько признаков есть в этом наборе данных.
Посчитайте, как много сериалов и фильмов представлено в этом датасете.
Посчитайте, как много пропущенных значений (NaN) есть в признаках director и country.
Посчитайте, сколько фильмов выпустили раньше 2003 года.
Посчитайте, какая страна чаще всего выпускает фильмы на Netflix.

```python
#Считывание датасета
films_data = pd.read_csv('netflix_titles.csv', sep = ',')
display(films_data)

```

```python
#информация о признаках и наблюдениях
print('Info :')
print(films_data.info())
#информация о числе сериалов и фильмов
print('TV_Shows and Films count:')
print(films_data["type"].value_counts())
```

```python
#подсчет пропусков в полях director и country
Na_country = films_data['country'].isnull().value_counts()[True]
print(f'Пропусков в поле country: {Na_country}')
Na_director = films_data['director'].isnull().value_counts()[True]
print(f'Пропусков в поле director: {Na_director}')
```

```python
#подсчет фильмов до 2003
under_2003 = films_data.loc[(films_data['release_year'] < 2003) & (films_data['type'] == 'Movie')].shape[0]
print(f'фильмов до 2003 = {under_2003}')
product_country = films_data.loc[(films_data['type'] == 'Movie')]['country'].value_counts().index[0]
print(f'выпускала больше всего {product_country}')
```
